```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pg = 6.5
mux = pg
sigmax = 0.10*mux
sg = np.random.normal(mux, sigmax, 1000)
```

```python
plt.hist(sg, bins=20)
plt.ylabel("Probabilidade")
plt.xlabel("Dados")
plt.title("Histograma Carga $g_{k}$")
plt.savefig('CargaGk.svg', dpi=600, bbox_inches="tight", transparent=True)
```

```python
pq = 3.5
mux = 0.93*pq
sigmax = 0.20*mux
sq = np.random.normal(mux, sigmax, 1000)
```

```python
plt.hist(sq, bins=20)
plt.ylabel("Probabilidade")
plt.xlabel("Dados")
plt.title("Histograma Carga $q_{k}$")
plt.savefig('CargaQk.svg', dpi=600, bbox_inches="tight", transparent=True)
```
